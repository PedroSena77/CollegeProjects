Crie o vetor reclamacoes com todos os dados que cuja coluna Tipo seja igual a "Reclamação".


```python
reclamacoes = df.loc[df.loc[:,'Tipo'] == 'Reclamação']
reclamacoes
```

Mostre o número total de solictiações QtdeSolic de reclamações para cada tipo de serviço Servico


```python
typesReclamacoes = reclamacoes.groupby(['Servico'])['QtdeSolic'].sum()
typesReclamacoes
```

Quais são os maiores ou maior motivo de solicitação de reclamações?


```python
motivos = typesReclamacoes.loc[(typesReclamacoes == typesReclamacoes.min()) | (typesReclamacoes == typesReclamacoes.max())]

print(f'- Maior motivo: {motivos.keys()[0]}')
print(f'- Menor motivo: {motivos.keys()[1]}')
```
