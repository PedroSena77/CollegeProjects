14. Imprima o número total de dados ausentes em cada uma das linhas?

```python
values = df.isnull().values
emptylines = {}

i=0
for line in values:
    emptylines[i] = line.sum()
    i += 1

for line in emptylines:
    print(f'Quantidade de dados ausentes na linha {line}: {emptylines[line]}')
```
